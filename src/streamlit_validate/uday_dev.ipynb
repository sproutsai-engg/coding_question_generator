```python
import json
import os


with open('coding_questions_merged.json') as f:
    all_ques = json.load(f)
all_ques['1']['test_cases']['inputs']
```

```python
''' Ensuring the input test cases are as expected
 1. Converting dicts to lists
 2.Modifying the way strings are stored
'''
for ques in all_ques.keys():
    test_cases=all_ques[ques]['test_cases']
    inputs=test_cases['inputs']
    inp_list=[]
    # print(f'Ques is {ques}')
    for i in range(len(inputs)):
        if isinstance(inputs[i],dict):
            temp_list=[val for val in inputs[i].values()]
            inp_list.append(temp_list)
            # print(inp_list)

        elif isinstance(inputs[i],str):
            if inputs[i].startswith("`") and inputs[i].endswith("`"):
                inputs[i]=inputs[i][1:-1]
            try:
                parsed_inp=eval(inputs[i])
                if isinstance(parsed_inp,list):
                    inp_list.append(parsed_inp)
                elif isinstance(parsed_inp,tuple):
                    inp_list.append([parsed_inp])
                else:
                    inp_list.append(inputs[i])
            except (SyntaxError, NameError, TypeError) as e:
                print(f"Error evaluating input in {ques}'{inputs[i]}': {e}")
                inp_list.append(inputs[i])                 
        elif isinstance(inputs[i],int):
            inp_list.append(inputs[i])

        elif isinstance(inputs[i],list):
            inp_list.append(inputs[i])
    all_ques[ques]['test_cases']['inputs']=inp_list
    
```
